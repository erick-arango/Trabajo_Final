```python
import random
import os
import pandas as pd

# Función para cargar nombres desde un DataFrame
def cargar_nombres(dataframe, columna):
    return dataframe[columna].tolist()

# Función para distribuir nombres en semestres
def distribuir_nombres(nombres):
    if (len(nombres) != 1000):
        raise ValueError("La lista debe contener exactamente 1000 nombres.")

    semestre_1 = nombres[:140]
    semestre_2 = nombres[140:270]
    semestre_3 = nombres[270:390]
    semestre_4 = nombres[390:500]
    semestre_5 = nombres[500:600]
    semestre_6 = nombres[600:700]
    semestre_7 = nombres[700:790]
    semestre_8 = nombres[790:870]
    semestre_9 = nombres[870:940]
    semestre_10 = nombres[940:1000]

    return semestre_1, semestre_2, semestre_3, semestre_4, semestre_5, semestre_6, semestre_7, semestre_8, semestre_9, semestre_10

# Función para calcular horas según los créditos
def calcular_horas(creditos):
    if creditos == 1:
        return "HTD:16 horas y HTI:32 horas"
    elif creditos == 2:
        return "HTD:32 horas y HTI:64 horas"
    elif creditos == 3:
        return "HTD:64 horas y HTI:80 horas"
    elif creditos == 4:
        return "HTD:96 horas y HTI:120 horas"
    elif creditos == 12:
        return "Estas en tu semestre de practica"
    else:
        return "Créditos no válidos"

# Materias de cada semestre
materias_semestre_1 = {"Calculo diferencial": 3, "Algebra y Trigonometría": 3, "Geometría Vectorial y Analítica": 3, "Introducción a la ingeniería industrial": 1, "Lectoescritura": 3, "Vivamos la Universidad": 1, "Inglés 1": 1}
materias_semestre_2 = {"Calculo Integral": 3, "Algebra Lineal": 3, "Descubriendo la Física": 3, "Gestion de las organizaciones": 3, "Habilidades gerenciales": 3, "Inglés 2": 1}
materias_semestre_3 = {"Física Mecánica": 3, "Gestión Contable": 3, "Teoría general de sistemas": 3, "Probabilidad e Inferencia Estadística": 3, "Algoritmia y Programación": 3, "Inglés 3": 1}
materias_semestre_4 = {"Electiva en Física": 3, "Ingeniería Económica": 3, "Gestión de Métodos y Tiempos": 4, "Diseño de Experimentos y Análisis de Regresión": 3, "Optimización": 3, "Inglés 4": 1}
materias_semestre_5 = {"Gestión Financiera": 3, "Dinámica de Sistemas": 3, "Gestión por Procesos": 3, "Muestreo y Series de Tiempo": 3, "Procesos Estocásticos y Análisis de Decisión": 3, "Laboratorio Integrado de Física": 1, "Ingles 5": 1}
materias_semestre_6 = {"Formación Ciudadana y Constitucional": 1, "Gestión Tecnológica": 3, "Normalización y Control de Calidad": 3, "Formulación de Proyectos de Investigación": 3, "Simulación Discreta": 3, "Electiva en Humanidades": 3, "Ingles 6": 1}
materias_semestre_7 = {"Legislación": 3, "Emprendimiento": 2, "Diseño de Sistemas Productivos": 3, "Énfasis Profesional 1": 3, "Electiva Complementaria 1": 3, "Electiva en Humanidades 2": 3}
materias_semestre_8 = {"Formulación y Evolución de Proyectos de Inversión": 3, "Administración de la Producción y del Servicio": 3, "Énfasis Profesional 2": 3, "Electiva Complementaria 2": 3, "Electiva en Humanidades 3": 3}
materias_semestre_9 = {"Gestión de proyectos": 3, "Ingeniería del Mejoramiento Continuo": 3, "Gestión de la Cadena de Abastecimiento": 3, "Énfasis Profesional 3": 3, "Electiva Complementaria 3": 3, "Electiva en Humanidades 4": 3}
materias_semestre_10 = {"Práctica Profesional": 12}

# Lista con diccionarios de semestres
materias_semestres = [materias_semestre_1, materias_semestre_2, materias_semestre_3, materias_semestre_4, materias_semestre_5, materias_semestre_6, materias_semestre_7, materias_semestre_8, materias_semestre_9, materias_semestre_10]

# Lista con la cantidad de aulas para cada semestre
aulas_por_semestre = [5, 5, 4, 5, 4, 4, 5, 4, 4, 6]

# Cargar nombres y apellidos desde GitHub
nombres_df = pd.read_csv("https://raw.githubusercontent.com/erick-arango/Trabajo_Final/main/nombres.csv")
apellidos_df = pd.read_csv("https://raw.githubusercontent.com/erick-arango/Trabajo_Final/main/apellidos.csv")

# Verificar nombres de columnas y cargar datos
print("Columnas en el archivo de nombres:", nombres_df.columns)
print("Columnas en el archivo de apellidos:", apellidos_df.columns)

nombres_columna = nombres_df.columns[0]
apellidos_columna = apellidos_df.columns[0]

nombres = cargar_nombres(nombres_df, nombres_columna)
apellidos = cargar_nombres(apellidos_df, apellidos_columna)

# Genera 1000 nombres aleatorios
nombres_generados = []
for _ in range(1000):
    nombre_completo = random.choice(nombres) + ' ' + random.choice(nombres) + ' ' + random.choice(apellidos) + ' ' + random.choice(apellidos)
    nombres_generados.append(nombre_completo)

# Distribuir los nombres en los semestres
semestre_1, semestre_2, semestre_3, semestre_4, semestre_5, semestre_6, semestre_7, semestre_8, semestre_9, semestre_10 = distribuir_nombres(nombres_generados)

# Función para asignar estudiantes a materias
def asignar_estudiantes_a_materias(materias, estudiantes, cantidad_por_grupo, aulas, semestre):
    asignaciones = []
    for i, (materia, creditos) in enumerate(materias.items()):
        horas = calcular_horas(creditos)
        for j in range(1, aulas + 1):
            codigo = f"{materia[:3].upper()}{creditos}{i+1}{j}"
            start_index = (j - 1) * cantidad_por_grupo
            end_index = start_index + cantidad_por_grupo
            estudiantes_grupo = estudiantes[start_index:end_index]
            asignaciones.append({
                'Codigo': codigo,
                'Materia': materia,
                'Horas': horas,
                'Estudiantes': estudiantes_grupo,
                'Semestre': semestre
            })
    return asignaciones

# Asignar estudiantes a materias para cada semestre
cantidad_por_grupo_dict = {1: 30, 2: 30, 3: 30, 4: 25, 5: 25, 6: 25, 7: 20, 8: 20, 9: 20, 10: 10}

# Crear carpetas principales
os.makedirs("trabajo_final", exist_ok=True)
os.makedirs("semestres", exist_ok=True)

for i, (materias, estudiantes) in enumerate(zip(materias_semestres, [semestre_1, semestre_2, semestre_3, semestre_4, semestre_5, semestre_6, semestre_7, semestre_8, semestre_9, semestre_10]), start=1):
    cantidad_por_grupo = cantidad_por_grupo_dict[i]
    aulas = aulas_por_semestre[i-1]
    asignaciones = asignar_estudiantes_a_materias(materias, estudiantes, cantidad_por_grupo, aulas, i)

    # Crear carpeta para el semestre
    semestre_folder = f"semestres/Semestre_{i}"
    os.makedirs(semestre_folder, exist_ok=True)

    # Guardar las asignaciones en archivos de Excel y CSV por materia
    for asignacion in asignaciones:
        materia_folder = os.path.join(semestre_folder, asignacion['Materia'])
        os.makedirs(materia_folder, exist_ok=True)

        # Guardar en Excel
        df = pd.DataFrame(asignacion['Estudiantes'], columns=['Estudiantes'])
        df['Materia'] = asignacion['Materia']
        df['Codigo'] = asignacion['Codigo']
        df['Horas'] = asignacion['Horas']
        df['Semestre'] = asignacion['Semestre']
        excel_filename = os.path.join(materia_folder, f"{asignacion['Codigo']}_{len(asignacion['Estudiantes'])}.xlsx")
        df.to_excel(excel_filename, index=False)

        # Guardar en CSV
        csv_filename = os.path.join(materia_folder, f"{asignacion['Codigo']}_{len(asignacion['Estudiantes'])}.csv")
        df.to_csv(csv_filename, index=False)
```
